```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('C:/Users/12034/Downloads/addedGyroscope.csv')

df.head()
```

```python

df['Az_2'] = df.Az - 1.0
df['angle_accel_y'] = np.arctan( df.Ay / np.sqrt( df.Ax**2 + df.Az_2**2))
df['angle_accel_z'] = np.arctan( df.Az_2 / np.sqrt( df.Ax**2 + df.Ay**2))



df_filter = df[df['Timestamp']>1676841200]
plt.plot(df_filter.Timestamp, df_filter.angle_accel_z)
plt.plot(df_filter.Timestamp, df_filter.angle_accel_y)
# plt.plot(df_filter.Timestamp, df_filter.Az)


def dataToAngle(axisOut, axis2, axis3):
    return np.arctan(axisOut / np.sqrt( axis2**2 + axis3**2))
```
