```python
import pandas as pd
import json
import datetime
```

```python
def csvToLineProtocol(datapoint):
    now = datetime.datetime.now()
    timestamp = int(now.timestamp())

    rpm = float(datapoint['RPM'].strip().split(' ')[0])
    speed = float(datapoint['Speed'].strip().split(' ')[0])
    throttlePos = float(datapoint['Throttle POS'].strip().split(' ')[0])

    baseString = "telemetry,id=1,make=toyota,model=rav4,year=2022 RPM={},Speed={},ThrottlePosition={} {}"

    payload = baseString.format(rpm, speed, throttlePos, now)
    
    return payload
```

```python
df = pd.read_csv('2023-03-20_16-52-13_kia_optima_2015_vehicle_telemetry (1).csv')
df.columns = [x.strip() for x in df.columns]
```

```python
j = 100

datapoint = df.iloc[j]
```

```python
csvToLineProtocol(datapoint)
```
